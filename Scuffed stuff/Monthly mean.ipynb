## **Der hvor data er lig en værdi**

```python
def dato_lig(data, lig):
    indeks = []
    for i in range(len(data)):
        if data[i]==lig:
            indeks.append(i)
    count = 0
    print(f'Dataet={lig} {len(indeks)} gange, og sker ved måned(erne):')
    for i in range(len(indeks)):
        print(f'{måned[indeks[count]]} / {år[indeks[count]]}')
        count +=1

dato_lig(N, 0)
```
